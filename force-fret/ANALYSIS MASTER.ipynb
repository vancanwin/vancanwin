```python
#fitting force vs. fret with sigmoidal
def sigmoid_function(xdata, x0, k, c):
    #xdata = all your data
    y = c/(1+np.exp(-k*(xdata-x0)))
    return y

def param_fx(fret_high,fret_low,opening_force):
    if opening_force > 5:
        return fret_high, fret_low, opening_force

def residuals(p_guess, xdata, ydata):
    return ydata - sigmoid_function(xdata,p_guess[0],p_guess[1],[])

def sigmoid_derivative(xdata,x0,k):
    f = np.exp(-k*(xdata-x0))
    return -k/f
```
