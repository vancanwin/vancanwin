```python
# Pulling all files from the directory and only grabbing the ".str" files
arrAllFiles = os.listdir('.')
arrStrFiles = []
for file in arrAllFiles:
    if file[len(file)-4:len(file)] == '.str':
        arrStrFiles.append(file)
    
# Cycling through each of the .str files        
# Uploading the data into the database
arrInserts = []
for file in arrStrFiles:
    # Marking the various attributes based on the filename
    file_attributes = file.strip(".str").split("_")
    if len(file_attributes) == 3:
        file_attributes.append(1)
    if len(file_attributes) < 4:
        raise Exception("This file %s doesnt have enough attributes (4)" % (file))
       
    if concentration != file_attributes[1] or molecule != file_attributes[2]:
        # Skipping anything that didn't have the right concentration or molecule
        continue
    
    file_indicator = file_attributes[3]
    
    # Opening that file and creating a csv reader from that
    f = open(file, 'r')
    csv_reader = csv.reader(f, delimiter = '\t')
    
    # Pulling out the stiffness, step size, and sampling time from the header lines
    stiffness_line = csv_reader.next()[0]
    stiffness = stiffness_line[stiffness_line.find('(')+1:stiffness_line.find(')')].split(",")
    step_line = csv_reader.next()[0]
    step_size = step_line[step_line.find("=") + 1:len(step_line)]
    sampling_line = csv_reader.next()[0]
    sampling_time = sampling_line[sampling_line.find("=") + 1:len(sampling_line)]
    
    dummy = csv_reader.next() # header row
    
    # Looping through the rest of the .str files
    cnt_line = 0
    for line in csv_reader:
        cnt_line += 1
        arrInserts.append([concentration, molecule, file_indicator, stiffness[0], stiffness[1], step_size, sampling_time, cnt_line, line[0], line[1], line[2]])
    
    # Uploading the data into the database
    if len(arrInserts) > 0:
        conn.executemany("insert into %s values (%s)" % (saveTable, ','.join([ '%s' ] * len(arrInserts[0]))), arrInserts)
        arrInserts = []
        dbConn.commit()
```
