```python
from czifile import CziFile
import czifile
import pandas as pd
import numpy as np
import scanpy, cv2,os
import matplotlib.pyplot as plt
import matplotlib as mpl
import logging


def OpenCZI(fname, analysis_filename):
    mpl.rcParams['figure.dpi'] = 50

    channels = []
    color = 1
    rho_theta_df = pd.DataFrame([])
    initial_list = {"rho":[0],"theta":[0]}
    rho_theta_list= pd.DataFrame(initial_list)
    count_img = 0
    dpi = 200
    for stack in range(img2.shape[4]):
        img = np.zeros((img2.shape[5],img2.shape[6],3), np.uint8)
        channel_raster = img2[0,0,color,0,stack,:,:,0] #select only one channel
        channels.append(channel_raster)
        arr = np.stack(channels, axis=2)
      
        img[:,:,0] = arr[:,:,stack]
        img[:,:,1] = arr[:,:,stack]
        img[:,:,2] = arr[:,:,stack]

        images = []
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        images.append(gray)

        kernel_size = 5
        blur_gray = cv2.GaussianBlur(gray,(kernel_size, kernel_size),0)

        #strict
        # low_threshold = 225
        # high_threshold = 200
        
        #loose
        low_threshold = 100
        high_threshold = 200
        edges = cv2.Canny(blur_gray,low_threshold,high_threshold,apertureSize = 3)
        images.append(edges)


        rho = 1  # distance resolution in pixels of the Hough grid
        theta = 5*np.pi / 180  # angular resolution in radians of the Hough grid
        threshold = 15  # minimum number of votes (intersections in Hough grid cell)
        min_line_length = 10#20  # minimum number of pixels making up a line
        max_line_gap = 5 # maximum gap in pixels between connectable line segments
        line_image = np.copy(img) * 0  # creating a blank to draw lines on

        # Run Hough on edge detected image
        # Output "lines" is an array containing endpoints of detected line segments
        rho_theta = cv2.HoughLines(edges, rho, theta, threshold, np.array([]),
                            min_line_length, max_line_gap)
       
        try:
            print("\nfirst item: ", len(rho_theta), "rho:",rho_theta[0])
            go_no_go = "yes"
           
                  
        except:
            go_no_go = "no"
            pass
        
        if go_no_go == "yes":
            
            for i in enumerate(rho_theta):
                iterate_num = i[0]
                rho_theta_values = {"rho":[rho_theta[iterate_num][0][0]],
                                    'theta':[rho_theta[iterate_num][0][1]]}
                df2 = pd.DataFrame(rho_theta_values)

                rho_theta_list = pd.concat((rho_theta_list,df2),axis=0)
        
        
            

        print("list", rho_theta_list)     
        lines = cv2.HoughLinesP(edges, rho, theta, threshold, np.array([]),
                            min_line_length, max_line_gap)     

        if lines is None:
            pass
        else:
            
            for line in lines:
                for x1,y1,x2,y2 in line:
                    cv2.line(line_image,(x1,y1),(x2,y2),(255,0,0),5)

            lines_edges = cv2.addWeighted(img, 0.8, line_image, 1, 0)
            images.append(lines_edges)

            #display images
            x = 20
            f, axarr = plt.subplots(1,3,figsize=(2*x,x))
            plt.xticks([])
            plt.yticks([])
            plt.xlabel(' ')
            plt.ylabel(' ')
            axarr[0].imshow(images[0])
            axarr[1].imshow(images[1])
            axarr[2].imshow(images[2])
            plt.subplots_adjust(wspace=0, hspace=0)
            plt.axis('off')

            plt.show()
            f.savefig("_".join([analysis_filename,str(count_img)+'.png']),dpi=dpi)
            count_img +=1

    
    rho_theta_list.reset_index(drop=True)
    rho_theta_list['theta'] *= 180/np.pi
    rho_theta_list[rho_theta_list['theta']>180] -=180 
    
    num_data_points =rho_theta_list.shape[0]
    sqrt_num_data_points = np.sqrt(num_data_points)
    max_data_points = rho_theta_list['theta'].max()
    min_data_points = rho_theta_list['theta'].min()
    rho_theta_list['theta_normalize'] = rho_theta_list['theta']/max_data_points
    binsize = int(np.ceil(sqrt_num_data_points)) #100
    
    f2 = plt.figure(figsize=(2*x,x)) 
    # hist = rho_theta_list['theta_normalize'].hist(bins= binsize)
    hist = rho_theta_list['theta'].hist(bins= binsize)
    f2.savefig("_".join([analysis_filename,'hist.png']),dpi=dpi)
    
    rho_theta_list.to_csv(analysis_filename+"rho_theta_list.csv")
    return rho_theta_list

#open folder
image_archive_folder =  'C:\\Users\\Van\\Downloads\\20221115\\'
os.chdir(image_archive_folder)
folder_list = os.listdir(image_archive_folder)[:-1]

for folder in folder_list:
    img_folder = image_archive_folder+folder+"\\"
    os.chdir(img_folder)
    img_files = ["Z1","Z2"]
    for img in img_files:
        fname = img_folder+img+".czi"
        img2 = czifile.imread(fname)
        analysis_filename = folder+img
        df = OpenCZI(img,analysis_filename)
        print(df)
```
